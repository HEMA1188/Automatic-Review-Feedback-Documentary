```python
import transformers
import torch

# Load the BART model
model_name = "facebook/bart-base"
tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)
model = transformers.AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Generate a reply to the review
def generate_reply(review_text):
    input_ids = tokenizer.encode("Review: " + review_text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate the reply
    outputs = model.generate(input_ids, max_length=256, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.92)

    reply = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return reply

# Example usage:

review_text = "Great app! Love it!"
reply = generate_reply(review_text)

print(reply)
```

```python
import transformers
import torch

# Load the BART model
model = transformers.AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-base")

# Generate a reply to the review
def generate_reply(review_text):
    input_ids = transformers.AutoTokenizer.from_pretrained("facebook/bart-base")(review_text, return_tensors="pt")["input_ids"]

    # Generate the reply
    outputs = model.generate(input_ids, max_length=256, num_beams=4, length_penalty=2.0, early_stopping=True)
    reply = transformers.AutoTokenizer.from_pretrained("facebook/bart-base").decode(outputs[0], skip_special_tokens=True)
    return reply

# Example usage:

review_text = "The app is okay, but it could be better."
reply = generate_reply(review_text)

print(reply)
```
